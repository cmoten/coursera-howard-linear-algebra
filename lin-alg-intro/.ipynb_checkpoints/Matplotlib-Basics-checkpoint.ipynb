# Matplotlib Basics

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Basic example

```python
x = np.linspace(-3,3,20)
y = x**2
x
```

```python
plt.plot(x,y,'b-')
```

## Basic Matplotlib commands

```python
plt.plot(x,y,'g*-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("First Plot")
```

## Creating multiple plots on the same canvas

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'g')
plt.title("Y is a function of X")

plt.subplot(1, 2, 2)
plt.plot(y,x,'r--')
plt.title("X is a function of Y")
```

## Introduction to object-oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.8])
axes.plot(x,y,'b*')
```
